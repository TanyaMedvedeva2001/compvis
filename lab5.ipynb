```python
import cv2
import numpy as np
from google.colab import drive
from google.colab.patches import cv2_imshow
drive.mount('/content/gdrive')
im1 = cv2.imread('gdrive/My Drive/lab5_1.jpg')
im1 = cv2.resize(im1, (300, 500))
im1 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
im2 = cv2.imread('gdrive/My Drive/lab5_2.jpg')
im2 = cv2.resize(im2, (300, 500))
im2 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
im3 = cv2.imread('gdrive/My Drive/lab5_3.jpg')
im3 = cv2.resize(im3, (300, 500))
im3 = cv2.cvtColor(im3, cv2.COLOR_BGR2GRAY)

```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

faces1 = face_cascade.detectMultiScale(im1)
faces2 = face_cascade.detectMultiScale(im2)
faces3 = face_cascade.detectMultiScale(im3)
for x, y, width, height in faces1:
    cv2.rectangle(im1, (x, y), (x + width, y + height), color=(255, 0, 0), thickness=2)

for x, y, width, height in faces2:
    cv2.rectangle(im2, (x, y), (x + width, y + height), color=(255, 0, 0), thickness=2)

for x, y, width, height in faces3:
    cv2.rectangle(im3, (x, y), (x + width, y + height), color=(255, 0, 0), thickness=2)
```

```python
cv2_imshow(im1)
cv2_imshow(im2)
cv2_imshow(im3)
```
