```python
import cv2
from google.colab import drive
from google.colab.patches import cv2_imshow
drive.mount('/content/gdrive')
img = cv2.imread('gdrive/My Drive/lab3.jpg')
img = cv2.resize(img, (400, 400))
cv2_imshow(img)

```

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB, 1)
l, a, b = cv2.split(lab)

l2 = cv2.equalizeHist(l)

lab2 = cv2.merge((l2,a,b))
img2 = cv2.cvtColor(lab2, cv2.COLOR_LAB2BGR)
cv2_imshow(img2)
```

```python
cv2_imshow(cv2.blur(img, (5,5)))
```

```python
cv2_imshow(cv2.GaussianBlur(img, (5, 5), 0))
```

```python
import numpy as np
img_grey = cv2.imread('gdrive/My Drive/lab3.jpg', cv2.IMREAD_GRAYSCALE)
img_grey = cv2.resize(img_grey, (400, 400))
kernel = np.ones((3, 3), np.uint8)
cv2_imshow(cv2.morphologyEx(img_grey, cv2.MORPH_OPEN, kernel))
```

```python
cv2_imshow(cv2.morphologyEx(img_grey, cv2.MORPH_CLOSE, kernel, iterations=1))
```
