```python
import cv2
from google.colab import drive
from google.colab.patches import cv2_imshow
drive.mount('/content/gdrive')
img = cv2.imread('gdrive/My Drive/lab2.jpg')
img = cv2.resize(img, (400, 400))
cv2_imshow(img)

lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB, 1)
l, a, b = cv2.split(lab) 
a2 = a + 10
b2 = b + 10
a3 = a - 10
b3 = b - 10
lab2 = cv2.merge((l,a2,b2))
img2 = cv2.cvtColor(lab2, cv2.COLOR_LAB2BGR)
cv2_imshow(img2)
lab3 = cv2.merge((l,a3,b3))
img3 = cv2.cvtColor(lab3, cv2.COLOR_LAB2BGR) 
cv2_imshow(img3)
```
